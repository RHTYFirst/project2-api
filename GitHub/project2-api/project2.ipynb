```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links

import requests
from bs4 import BeautifulSoup

# Step 1: Define the URL of the Wikipedia page (Formula One)
url = "https://en.wikipedia.org/wiki/Formula_One"

# Step 2: Make a request to fetch the HTML content of the page
response = requests.get(url)

# Step 3: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 4: Find all the links (anchor tags with href)
links = soup.find_all('a', href=True)

# Step 5: Loop through the first 10 links
for i, link in enumerate(links[:10]):
    # Construct the full URL if it's a relative path
    link_url = link['href']
    if not link_url.startswith('http'):
        link_url = 'https://en.wikipedia.org' + link_url

    # Fetch the linked page's content
    link_response = requests.get(link_url)
    link_soup = BeautifulSoup(link_response.text, 'html.parser')

    # Get the text from the page and print the first 100 characters
    link_text = link_soup.get_text()[:100]
    print(f"Link {i+1}: {link_url}")
    print(f"First 100 characters of content:\n{link_text}\n")
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

import requests
from bs4 import BeautifulSoup


script_texts = []


for i, row in scripts.iterrows():
    url = row['link']
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, 'html.parser')
        
        clean_text = soup.get_text()
        
        first_1000_chars = clean_text[:1000]
        
        script_texts.append(first_1000_chars)
        print(f"Processed: {url}")
        
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve {url}: {e}")
        script_texts.append("Failed to retrieve content")

```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"

import pandas as pd

scripts['text'] = script_texts

scripts.to_csv("pudding_texts.csv", index=False)
print("Saved to pudding_texts.csv")
```

```python
import requests

# Loop over the top 10 movies and download their posters
for i, row in dfTMDBtop10.iterrows():
    poster_url = row['url']
    
    # Fetch the poster image
    poster_response = requests.get(poster_url)
    
    # Save the poster as a file in the repository
    if poster_response.status_code == 200:
        with open(f"{row['title'].replace('/', '_')}.jpg", 'wb') as file:
            file.write(poster_response.content)
        print(f"Downloaded: {row['title']}")
    else:
        print(f"Failed to download: {row['title']}")
```
